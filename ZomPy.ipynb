```python
# Simple pygame program
# Import and initialize the pygame library
import pygame
import os
import random
from pygame.locals import *
```

```python
ALPHA = (0,255,0)
```

```python
pygame.init()

punti = 0
# Set up the drawing window (input take a list or a tuple)
window = (1600, 800)
screen = pygame.display.set_mode(window)

# background image
bg = pygame.image.load(os.path.join('textures', 'bg.jpg')) #.convert()
bgX = 0 # x position of first image
bgX2 = bg.get_width() # x position of second image (starts when the first one ends)
# Milos
#Milos = pygame.Surface((150, 160))
Milos = pygame.image.load(os.path.join('textures', 'milos_walk.jpg')).convert()
Milos.convert_alpha()
Milos.set_colorkey(ALPHA)
rect = Milos.get_rect()
x_milos = 100
y_milos = 500

#Ostacolo
Ostacolo = pygame.Surface((100,200))
ost_x = bg.get_width()

def spawn():
    global ost_x
    if ost_x < 0:
        ran = random.uniform(0, 1)
        if ran < 0.005:
           ost_x = 1500 
        
def zompare():
    global y_milos
    y_milos -=400
    screen.blit(Milos, (x_milos, y_milos))

def gravity():
    global y_milos
    if y_milos < 500:
        y_milos += 1
        screen.blit(Milos, (x_milos, y_milos))  

# clock setting
clock = pygame.time.Clock()

def updateScreen():
    screen.blit(bg, (bgX, 0))
    screen.blit(bg, (bgX2, 0))
    screen.blit(Ostacolo, (ost_x,500))
    
    if y_milos < 500:
        gravity()
    else:
        screen.blit(Milos, (x_milos, y_milos))
        
    pygame.display.update()
    
    # update vs flip:
    # flip updates ALL screen
    # update only updates part of the screen.
    #   called with no arguments is the same as flip

# custom events
increaseSpeed = pygame.USEREVENT + 1

# trigger custom event every 500 ms
pygame.time.set_timer(increaseSpeed, 500)

speed = 30

# Run until the user asks to quit
running = True
while running:

    updateScreen()
    spawn()
    if (x_milos+75) >= ost_x and (x_milos+75) <= ost_x + 100:
        if (y_milos + 160) >= 500:
            running = False
            
    bgX -= 1.4
    bgX2 -= 1.4
    ost_x -= 1.4
    if bgX < bg.get_width() * -1:
        bgX = bg.get_width()
    if bgX2 < bg.get_width() * -1:
        bgX2 = bg.get_width()

    # Handles events within the game execution
    for event in pygame.event.get():

        # if the user wants to quit
        if event.type == pygame.QUIT:
            running = False
        if event.type == KEYDOWN:
            if event.key == K_SPACE:
                if not y_milos < 500: 
                    zompare()
            if event.key == K_ESCAPE:
                running = False
        
        if event.type == increaseSpeed:
            speed += 10

    # tick the game clock at "speed" variable intervals
    clock.tick(speed)
    punti += 0.03

# Done! Time to quit.
pygame.quit()
```
